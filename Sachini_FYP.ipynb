```python

from bs4 import BeautifulSoup

def extract_stack_trace(html):
    soup = BeautifulSoup(html, 'html.parser')
    code_snippets = [code_tag.text for code_tag in soup.find_all('code')]
    filtered_code = [code for code in code_snippets if any(ext in code for ext in (".py", ".java", ".js", ".cs", ".cpp", ".go", ".rs", ".dart"))]
    return "\n".join(filtered_code)
```

```python
temp_df['stack_trace'] = temp_df['body'].apply(extract_stack_trace)
```

```python
import re
# def clean_html(text):
#     """Remove HTML tags and extra whitespace (with debugging)"""
#     # Track original type
#     original_type = type(text)

#     # Process only strings
#     if isinstance(text, str):
#         cleaned = BeautifulSoup(text, "lxml").get_text(separator=' ', strip=True)
#         cleaned = re.sub(r'\s+', ' ', cleaned)
#         cleaned = re.sub(r'&#x\d+;', '', cleaned)
#         result = cleaned.strip()
#     else:
#         result = text  # Return non-strings as-is

#     if not isinstance(text, str):
#         print(f"Skipped non-string: {original_type} → {text}")

#     return result

def clean_text(text):
    text = text.lower()  # Convert to lowercase
    text = re.sub(r'<[^>]+>', '', text)  # Remove HTML tags
    text = re.sub(r'[^a-z\s]', '', text)  # Keep only letters and spaces
    text = re.sub(r'\s+', ' ', text).strip()  # Remove extra spaces
    return text
```

```python
df_with_stack_trace["body"] = df_with_stack_trace['body'].apply(clean_text)
df_with_stack_trace["stack_trace"] = df_with_stack_trace['stack_trace'].apply(clean_text)
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")

def lemmatize_text(text):
    doc = nlp(text)
    return " ".join([token.lemma_ for token in doc if not token.is_stop])
```

```python
df_cleaned['title']=df_cleaned['title'].apply(lemmatize_text)
df_cleaned['body']=df_cleaned['body'].apply(lemmatize_text)
df_cleaned['stack_trace']=df_cleaned['stack_trace'].apply(lemmatize_text)
```

```python
import numpy as np
import gensim.downloader as api

glove_model = api.load("glove-wiki-gigaword-50")  # Load GloVe 50D embeddings

def get_embedding(text):
    tokens = text.split()
    vectors = [glove_model[word] for word in tokens if word in glove_model]
    return np.mean(vectors, axis=0) if vectors else np.zeros(50)  # Average word embeddings

pre_processed_data['embedded_title']=pre_processed_data['title'].apply(get_embedding)
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

mlb = MultiLabelBinarizer()
y = mlb.fit_transform(pre_processed_data['tags'])
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(pre_processed_data['embedded_title'].tolist(), y, test_size=0.2, random_state=42, stratify=y)
```
